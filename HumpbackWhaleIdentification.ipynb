# Humpback Whale Detection using CNN

Solution for the kaggle challenge: https://www.kaggle.com/c/humpback-whale-identification/data

```python
import pandas as pd
import numpy as np
import tensorflow as tf

%matplotlib inline
```

```python
table = pd.read_csv('..\\data\\HumpbackWhaleDetection\\train.csv')
```

```python
table.head()
```

```python
nClasses = len(table["Id"].unique())
```

```python
classes = table["Id"].unique().tolist()
classes
```

```python
classesDict = {classes[i]:i for i in range(0, len(classes))}
```

```python
classesDict
```

```python
import cv2
import os
from PIL import Image
```

```python
images = []
ids = []
```

```python
len(table)
```

```python
table.Image[25360]
```

```python
for i in range(0, len(table)):
    img = cv2.imread('..\\data\\HumpbackWhaleDetection\\trainingSet\\'+table.Image[i], cv2.IMREAD_COLOR)
    imArr = Image.fromarray(img, 'RGB')
    imArr = imArr.resize((50, 50))
    imArr_1 = imArr.rotate(45)
    imArr_2 = imArr.rotate(75)
    imArr_3 = cv2.blur(np.array(imArr), (5, 5))
     
    images.append(np.array(imArr))
    images.append(np.array(imArr_1))
    images.append(np.array(imArr_2))
    images.append(np.array(imArr_3))
       
    for j in range(0,4):
        ids.append(classesDict[table.Id[i]])
    
    print('.')
        
```

```python
len(images)
```

```python
len(ids)
```

```python
images = np.array(images)
ids = np.array(ids)
```

```python
images.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(images, ids, test_size=0.15, random_state=42)
```

```python
print('Training set images length: '+str(len(X_train)))
print('Training set labels length: '+str(len(y_train)))
print('Test set images length: '+str(len(X_test)))
print('Test set labels length: '+str(len(y_test)))
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D, ZeroPadding2D, BatchNormalization
```

```python
model = Sequential()
model.add(Conv2D(16, (28, 28), strides=1, input_shape=X_train.shape[1:], activation=tf.nn.tanh, padding='valid'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(MaxPooling2D(pool_size=(2, 2),strides=1))

model.add(Conv2D(28, (13, 13), strides=2, activation=tf.nn.tanh, padding='same'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(MaxPooling2D(pool_size=(6, 6),strides=1))

model.add(Flatten())
model.add(Dense(750, activation=tf.nn.relu))
model.add(Dense(255, activation=tf.nn.relu))
model.add(Dropout(0.4))
model.add(Dense(500, activation=tf.nn.relu))
model.add(Dense(255, activation=tf.nn.relu))
model.add(Dropout(0.2))
model.add(Dense(500, activation=tf.nn.relu))
model.add(Dropout(0.2))

model.add(Dense(nClasses, activation=tf.nn.softmax))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=5, batch_size=50)
```
